```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style = 'dark')
```

```python
fifa_players = pd.read_csv('/Users/shekharsingh/Documents/Big Data and Business Analytics/Analytics1/RStudio/FIFA_2019.csv',index_col=0)
```

```python
plt.figure(figsize=(18,10))
sns.countplot(fifa_players['Overall'], palette='rocket')
plt.show()
```

```python
plt.figure(1 , figsize = (18 , 10))
sns.countplot(x = 'Position' , data = fifa_players , palette = 'rocket' )
plt.title('Count Plot of Postions of player')
plt.show()
```

```python
plt.figure(figsize=(20, 15))
sns.regplot(x = 'Age' , y = 'Overall', data = fifa_players )
plt.title('Age vs Overall rating')
plt.show()
```

```python
plt.figure(figsize=(14,12))
sns.heatmap(fifa_players.corr(),linewidths=1)
plt.title('Dataset Heatmap')
plt.show()
```
